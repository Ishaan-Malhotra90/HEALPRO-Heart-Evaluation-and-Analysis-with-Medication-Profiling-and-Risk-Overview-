### Finding Theraputic Classes with max and min Habit forming medicine

```python
df['Habit Forming'].value_counts()
```

```python
habit_forming_counts = df[df['Habit Forming'] == 'Yes']['Therapeutic Class'].value_counts()
non_habit_forming_counts = df[df['Habit Forming'] == 'No']['Therapeutic Class'].value_counts()
max_habit_forming_class = habit_forming_counts.idxmax()
min_habit_forming_class = habit_forming_counts.idxmin()

plt.figure(figsize=(12, 6))
habit_forming_counts.plot(kind='bar', color='blue', label='Habit Forming')
non_habit_forming_counts.plot(kind='bar', color='green', label='Non-Habit Forming')
plt.title('Habit Forming Medicines by Therapeutic Class')
plt.xlabel('Therapeutic Class')
plt.ylabel('Number of Medicines')
plt.legend()
print("OBSERVATIONS")
print(f'Therapeutic class with the maximum habit-forming medicines: {max_habit_forming_class}')
print(f'Number of habit-forming medicines in {max_habit_forming_class}: {habit_forming_counts.max()}')
print(f'Therapeutic class with the minimum habit-forming medicines: {min_habit_forming_class}')
print(f'Number of habit-forming medicines in {min_habit_forming_class}: {habit_forming_counts.min()}')
plt.show()
```

### Analysis of Side effects with different columns

### Most commmon type of side effect in the dataset

```python
all_side_effects = pd.concat([df['sideEffect0'], df['sideEffect1'], df['sideEffect2']])
most_common_side_effect = all_side_effects.mode()[0]
print("OBSERVATIONS")
print(f'The most common side effect among SideEffect0, SideEffect1, and SideEffect2 is: {most_common_side_effect}')
```

### Most Common type of side effect according to Therapeutic Classes

```python
all_side_effects = pd.concat([df[['Therapeutic Class', 'sideEffect0']], df[['Therapeutic Class', 'sideEffect1']], df[['Therapeutic Class', 'sideEffect2']]])
most_common_side_effect_by_class = all_side_effects.groupby('Therapeutic Class')['sideEffect0'].agg(lambda x: x.mode().iat[0] if not x.empty else 'No Side Effects')
print("OBSERVATIONS")
print(most_common_side_effect_by_class)
```

### Most common type of side effect according to use

```python
data = pd.concat([df[['use0', 'sideEffect0']], df[['use0', 'sideEffect1']], df[['use0', 'sideEffect2']]])
most_common_side_effect_by_use = data.groupby('use0')['sideEffect0'].agg(lambda x: x.mode().iat[0] if not x.empty else 'No Side Effects')
print("OBSERVATIONS")
print(most_common_side_effect_by_use)
```

### Most common uses according to different Therapeutic Classes
