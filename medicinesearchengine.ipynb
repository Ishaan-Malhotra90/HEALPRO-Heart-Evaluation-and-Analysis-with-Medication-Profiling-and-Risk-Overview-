```python
therapeutic_classes_to_keep = ['CARDIAC', 'GASTRO INTESTINAL', 'BLOOD RELATED', 'PAIN ANALGESICS', 'ANTI DIABETIC']
mask = df['Therapeutic Class'].isin(therapeutic_classes_to_keep)
df = df[mask]
df
```

```python
df.reset_index(drop=True, inplace=True)
df
```

### Linear Search function for medicine search

```python
query = input('Enter the medicine name: ').lower()
result_list = []
for i, name in enumerate(df['name']):
    if query in name.lower():
        result_list.append(i)
if not result_list:
    print("No matches found.")
else:
    result_df = df.iloc[result_list]
    print("The medicines matched for your search are:-")
    for idx, (index, row) in enumerate(result_df.iterrows(), start=1):
        print("Index:", idx)
        print("Name:", row['name'])
        print("Substitutes:", row['substitute0'], row['substitute1'])
        print("Use:", row['use0'])
        print("Side Effect:", row['sideEffect0'], row['sideEffect1'], row['sideEffect2'])
        print("Therapeutic Class:", row['Therapeutic Class'])
        print("Habit Forming:", row['Habit Forming'])
        print()
```
